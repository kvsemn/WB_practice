### Классический Python ###

```python
products = ['автомобиль Haval', 'обувь', 'одежда', 'телефоны']
print(products)
```

```python
prices = [20000, 3000, 45600, 5000]
print(prices)
```

```python
def profit_python(sum):    
    for s in sales1:
        ps = prices[0]*s
        sum = sum + ps
    for s in sales2:
        ps = prices[1]*s
        sum = sum + ps
    for s in sales3:
        ps = prices[2]*s
        sum = sum + ps
    for s in sales4:
        ps = prices[3]*s
        sum = sum + ps
    return sum
```

```python
# Python program for implementation of heap Sort

def heapify(arr, N, i):
	largest = i 
	l = 2 * i + 1	
	r = 2 * i + 2	 

	if l < N and arr[largest] < arr[l]:
		largest = l

	
	if r < N and arr[largest] < arr[r]:
		largest = r

	if largest != i:
		arr[i], arr[largest] = arr[largest], arr[i] # swap

		# Heapify the root.
		heapify(arr, N, largest)



def heapSort(arr):
	N = len(arr)

	for i in range(N//2 - 1, -1, -1):
		heapify(arr, N, i)

	for i in range(N-1, 0, -1):
		arr[i], arr[0] = arr[0], arr[i] # swap
		heapify(arr, i, 0)
        
def print_sorted_arr(arr):
    N = len(arr)
    print("Sorted array is")
    for i in range(N):
        print("%d" % arr[i], end=" ")
        
    

```

```python
arr = [12, 11, 13, 5, 6, 7]
heapSort(arr)
print_sorted_arr(arr)
```

```python
%%timeit 
heapSort(arr)
```

```python
import numpy as np

def heapify_numpy(arr, n, i):
    largest = i  
    l = 2 * i + 1  
    r = 2 * i + 2  

    # Если левая дочерняя вершина больше корня
    if l < n and arr[i] < arr[l]:
        largest = l

    # Если правая дочерняя вершина больше самой большой вершины на данный момент
    if r < n and arr[largest] < arr[r]:
        largest = r

    # Если самая большая вершина не является корнем
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]  # Обмен корня и наибольшего элемента
        heapify_numpy(arr, n, largest)

def heapsort_numpy(arr):
    n = len(arr)

    # Построение max-heap
    for i in range(n // 2 - 1, -1, -1):
        heapify_numpy(arr, n, i)

    # Извлечение элементов по одному из кучи
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]  
        heapify_numpy(arr, i, 0)
```

```python
# Пример использования алгоритма heapsort, используя numpy
arr = np.array([12, 11, 13, 5, 6, 7])
print("Исходный массив:", arr)
heapsort_numpy(arr)
print("Отсортированный массив:", arr)
```

```python
%%timeit
heapsort_numpy(arr)
```

```python
import numpy as np
data = np.genfromtxt('tarantino.csv', delimiter=',', names=True, dtype=[('movie', 'U16'), ('type', 'U16'), ('word', 'U16'), ('minutes_in', 'f8')])
for _ in data:
    print(_)
```

```python
#число летальных исходов в таблице
np.sum(data['type'] == 'death')
```

```python
#число проклятий в таблице
np.sum(data['type'] == 'word')
```

```python
#Отношение числа проклятий и числа летальных исходов
np.sum(data['type'] == 'word') / np.sum(data['type'] == 'death')
```

```python
#Частота употребления конкретных проклятий
unique_values, counts = np.unique(data['word'], return_counts=True)

dt = np.zeros(
    61,
    dtype={
        "names": ["word", "frequency_of_use"],
        "formats": ["U16", "i4"]
    }
)
dt['word'] = unique_values
dt['frequency_of_use'] = counts

for _ in dt:
    print(_)
```

```python
#Распределение времни между проклятиями
hist, bins = np.histogram(data[data['type'] == 'word']['minutes_in'])
print('Значения гистограммы: ', hist)
print('Границы интервалов: ', bins)
```
